```python
last_events = train.groupby('game_session', sort=False).last().reset_index()
last_events['type_shifted'] = last_events['type'].shift(1)
last_events['title_shifted'] = last_events['title'].shift(1)


is_consecutive_assessment = last_events['type'].eq('Assessment') & last_events['type_shifted'].eq('Assessment')
consecutive_assessments = last_events[is_consecutive_assessment]
with_labels = pd.merge(consecutive_assessments, train_labels.drop('title', axis=1), how='left', on=on)
with_labels.head(50)
```

```python
test = read_from_clean('test.ftr')
last_events = test.groupby('game_session', sort=False).last().reset_index()
last_events['type_shifted'] = last_events['type'].shift(1)
last_events['title_shifted'] = last_events['title'].shift(1)


is_consecutive_assessment = last_events['type'].eq('Assessment') & last_events['type_shifted'].eq('Assessment')
consecutive_assessments = last_events[is_consecutive_assessment]
consecutive_assessments.head(50)
```

```python
last_events = train.groupby('game_session', sort=False).last().reset_index()

def func(df):
    return df.assign(
        type_shifted=df['type'].shift(1),
        title_shifted=df['title'].shift(1),
    )

last_events = last_events.groupby('installation_id', sort=False).apply(func).reset_index(drop=True)
is_consecutive_assessment = last_events['type'].eq('Assessment') & last_events['type_shifted'].eq('Assessment')
consecutive_assessments = last_events[is_consecutive_assessment]
with_labels = pd.merge(consecutive_assessments, train_labels.drop('title', axis=1), how='left', on=on)
with_labels.head(50)
```
